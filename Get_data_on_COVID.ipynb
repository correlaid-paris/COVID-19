# Get the data 
From Ourworldindata

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

```python
total_cases = pd.read_csv("https://covid.ourworldindata.org/data/total_cases.csv")
```

# Get the data 
From other datasource:
1. https://www.who.int/emergencies/diseases/novel-coronavirus-2019/situation-reports
2. https://ncov2019.live/data 

```python
import requests

# put the name of the website in brackets to dls to download the file
dls = ""
resp = requests.get(dls)

output = open('data.xls', 'wb')
output.write(resp.content)
output.close()
```

# Visualise data

```python
cases100_France = total_cases.query("France > 100")["France"].copy()
days100_France = np.arange(len(cases100_France))

fig, ax = plt.subplots(figsize=(10,7))
plt.plot(days100_France, cases100_France, label='France', linewidth=2, marker='o')
plt.plot(np.array([11, 12]), np.array([3661, 4499]), label='France (Le Monde)', linewidth=2, marker='o')
plt.axvline(9, color='red', linestyle='dashed')
plt.axvline(13, color='red', linestyle='dashed')
ax.set_title('Total number of cases: France', fontsize='large')
ax.set_xlabel('Number of days since 100 cases', fontsize='large')
ax.set_ylabel('Total number of cases', fontsize='large')
plt.legend()
plt.show()
```
